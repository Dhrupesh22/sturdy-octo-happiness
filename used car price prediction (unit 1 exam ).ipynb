```python
df.drop(columns=['Unnamed: 0','New_Price','Name'], axis=1 ,inplace=True)
```

```python
indx = df[df['Power'] == 'null bhp'].index
```

```python
indx.shape
```

```python
df.drop(indx, axis=0, inplace=True)
```

```python
df['Mileage'] = df['Mileage'].str.split(' ').str[0]
df['Engine'] = df['Engine'].str.split(' ').str[0]
df['Power'] = df['Power'].str.split(' ').str[0]
```

```python
df['Mileage'] = df['Mileage'].astype(float)
df['Engine'] = df['Engine'].astype(float)
df['Power'] = df['Power'].astype(float)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df['Mileage'] = df['Mileage'].fillna(df['Mileage'].median())
df['Engine'] = df['Engine'].fillna(df['Engine'].median())
df['Power'] = df['Power'].fillna(df['Power'].median())
df['Seats'] = df['Seats'].fillna(df['Seats'].median())
```

```python
df.isna().sum()
```

```python
df = pd.get_dummies(df,columns=['Fuel_Type','Transmission','Owner_Type'],drop_first=True)
```

```python
df.head()
```

```python
xtrain =  df.drop(['Location','Price'],axis=1)
```

```python
xtrain.head()
```

```python
y = df['Price']
```

```python
scl = StandardScaler()
x_scl_train = scl.fit_transform(xtrain)
```

```python
model = Sequential()
model.add(Dense(10,activation='relu',input_dim = 14,kernel_regularizer=tf.keras.regularizers.l2(0.05)))
#model.add(Dropout(0.2))
model.add(Dense(5,activation='relu',input_dim = 14,kernel_regularizer=tf.keras.regularizers.l2(0.05)))
#model.add(Dropout(0.2))
model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(loss='mse',optimizer=tf.keras.optimizers.Adam(0.001))
```

```python
history = model.fit(x_scl_train,ytrain,epochs=150)
```

```python
ypred = model.predict(xtrain)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(ypred,y)*100
```
